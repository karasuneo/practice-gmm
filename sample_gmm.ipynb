```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
# 
mu1, sigma1 = [2, 2], [[1, 0.5], [0.5, 1]]
mu2, sigma2 = [7, 7], [[1, -0.3], [-0.3, 1]]
mu3, sigma3 = [12, 2], [[1, 0.3], [0.3, 1]]

# 
rv1 = multivariate_normal(mean=mu1, cov=sigma1)
rv2 = multivariate_normal(mean=mu2, cov=sigma2)
rv3 = multivariate_normal(mean=mu3, cov=sigma3)

# 
weights = [1 / 3, 1 / 3, 1 / 3]

# 
x, y = np.mgrid[0:15:0.1, 0:10:0.1]
pos = np.dstack((x, y))

# GMM
z1 = rv1.pdf(pos)
z2 = rv2.pdf(pos)
z3 = rv3.pdf(pos)
z = weights[0] * z1 + weights[1] * z2 + weights[2] * z3

# 
fig, ax = plt.subplots()

# 
ax.contour(x, y, z1, colors="blue", linestyles="dotted")
ax.contour(x, y, z2, colors="green", linestyles="dotted")
ax.contour(x, y, z3, colors="red", linestyles="dotted")

# GMM
ax.contour(x, y, z, colors="black")

# 
ax.set_title("GMM with 3 Gaussian Distributions")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
