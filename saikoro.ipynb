```python
import numpy as np
import matplotlib.pyplot as plt

# 乱数生成器のシードを設定（再現性のため）
np.random.seed(0)

# サイコロを振る回数
num_rolls = 10000

# 2つのサイコロを振る
dice1 = np.random.randint(1, 7, num_rolls)
dice2 = np.random.randint(1, 7, num_rolls)

# 2つのサイコロの和を計算
sum_dice = dice1 + dice2

# ヒストグラムをプロット（確率密度関数）
plt.hist(sum_dice, bins=range(2, 14), density=True, edgecolor="black", align="left")
plt.title("2つのサイコロの和の確率密度関数")
plt.xlabel("2つのサイコロの和")
plt.ylabel("確率密度")
plt.xticks(range(2, 13))
plt.grid(True)
plt.show()
```
